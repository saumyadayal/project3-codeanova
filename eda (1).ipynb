```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

# Exploratory Data Analysis

This notebook explores customer behaviour and identifies the factors contributing to customer churn.

```python
df.head()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Churn',data=df,palette='Set2')

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")

plt.show()
```

### Observation

The dataset shows the overall distribution of customers who stayed versus customers who churned.

```python
plt.figure(figsize=(6,5))
sns.countplot(x='gender',hue='Churn',data=df,palette='coolwarm')

plt.title("Gender vs Churn")

plt.show()
```

### Observation

This visualization compares churn across male and female customers.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Contract',hue='Churn',data=df,palette='viridis')

plt.xticks(rotation=15)

plt.show()
```

### Observation

Contract type strongly influences customer retention.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Churn',
            y='MonthlyCharges',
            data=df,
            palette='Set3')

plt.show()
```

### Observation

Customers with higher monthly charges may exhibit different churn patterns.

```python
plt.figure(figsize=(8,5))

sns.histplot(data=df,
             x='tenure',
             hue='Churn',
             kde=True)

plt.show()
```

### Observation

Customer tenure helps identify long-term and short-term customer behaviour.

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.show()
```

### Observation

Correlation matrix helps identify relationships among numerical variables.

# Conclusion
